```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#we imported numpy to make array of lines in x and y then using plt we ploted its graph
x= np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y)
```

```python
#used to graph the points rather than lines 
#plt.scatter(x[:20],y[:20])
plt.scatter(x[::10],y[::10],color='red',linewidth=1000)
```
